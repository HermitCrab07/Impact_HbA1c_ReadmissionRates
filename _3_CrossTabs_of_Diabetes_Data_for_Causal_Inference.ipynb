```python
# Specify covariates and group column
cat_cov = ['race', 'diabetesMed', 'insulin', 'change', 'gender']
group_col = 'Group'

# Calculate cross-tabulation with percentages
for covariate in cat_cov:
    cross_tab = pd.crosstab(df[group_col], df[covariate], normalize='index') * 100  # Calculate percentages within groups
    print(f"\nGroup Percentages for {covariate}:")
    print(cross_tab.to_string(float_format="{:.2f}%".format))  # Format and print
```

```python
import scipy.stats as stats

covariates_duringencounter = ['num_lab_procedures', 'change', 'diabetesMed', 'insulin', 'num_procedures', 'num_medications', 'number_emergency', 'number_diagnoses', 'readmitted']

for covariate in covariates_duringencounter:
    # Separate data for Treatment and Control groups
    treatment_data = df[df['Group'] == 'Treatment'][covariate]
    control_data = df[df['Group'] == 'Control'][covariate]

    if pd.api.types.is_numeric_dtype(df[covariate]):  # Check if covariate is numeric (continuous)
        # Perform t-test
        t_statistic, p_value = stats.ttest_ind(treatment_data, control_data)
        print(f"T-test for {covariate}: t-statistic = {t_statistic:.2f}, p-value = {p_value:.3f}")
    else:  # Covariate is categorical
        # Create contingency table
        contingency_table = pd.crosstab(df['Group'], df[covariate])

        # Perform chi-square test
        chi2_statistic, p_value, dof, expected = stats.chi2_contingency(contingency_table)
        print(f"Chi-square test for {covariate}: chi2-statistic = {chi2_statistic:.2f}, p-value = {p_value:.3f}")

```
